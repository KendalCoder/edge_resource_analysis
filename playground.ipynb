````python
def get_vsn_data(vsn, start, end):
    return query(
        start=start,
        end=end,
        filter={
            "vsn": vsn.upper(),
            },
        bucket="grafana-agent"
    )
````

````python
completed_runs_ex.to_csv("jobs.csv", index=False)
````

````python

for plugin_name, instances in completed_runs_ex.groupby("plugin_name"):
    vsn = ""
    merged = pd.DataFrame()
    # logging.info(f'Scraping metrics for {plugin_name}')
    # for _, v in tqdm(instances.iterrows(), desc=plugin_name):
    for _, v in instances.iterrows():
        instance = v.k3s_pod_instance
        vsn = v.vsn
        started = v.timestamp.isoformat()
        completed = v.completed_at.isoformat()
        extended_started = v.timestamp - pd.to_timedelta(1, unit='m')
        extended_completed = v.completed_at + pd.to_timedelta(1, unit='m')
        extended_started = extended_started.isoformat()
        extended_completed = extended_completed.isoformat()

        print(f'querying metrics on {started}-{completed} for {instance}')
        print(f'querying metrics on {extended_started}-{extended_completed} for {instance}')
        # df = get_vsn_data(vsn, started, completed)
        df = get_vsn_data(vsn, extended_started, extended_completed)
        if len(df) < 1:
            print(f'No record found for {instance}')
            continue
        print(f'{len(df)} records found')
        
        # container-related metrics
        # CPU
        container_df = df[df["meta.container"]==plugin_name]
        container_cpu_metrics = container_df[container_df["name"]=="container_cpu_usage_seconds_total"]
        print(f'{len(container_cpu_metrics)} CPU records found')
        print(container_cpu_metrics)
        container_cpu_util = calculate_cpu_utilization_from_cpuseconds(container_cpu_metrics.copy(), started)
        
        # Memory (workingset + swap)
        container_mem_rss_metrics = container_df[container_df["name"]=="container_memory_rss"]
        container_mem_workingset_metrics = container_df[container_df["name"]=="container_memory_working_set_bytes"]
        print(f'{len(container_mem_workingset_metrics)} Memory workingset records found')
        print(container_mem_workingset_metrics)
        container_mem_workingset_metrics["value_rss"] = container_mem_rss_metrics["value"].values
        container_mem_workingset_metrics["mem"] = container_mem_workingset_metrics["value"] + container_mem_workingset_metrics["value_rss"]
        merged_instance = pd.merge_asof(container_cpu_util[["timestamp", "cpu"]], container_mem_workingset_metrics[["timestamp", "mem"]], on="timestamp")

        # GPU
        # NOTE: GPU measurement does not well reflect how plugins use it
        #       because we take the instantaneous value at measuring time
        # new = pd.merge_asof(container_cpu_util[["timestamp", "value"]], container_mem_workingset_metrics[["timestamp", "mem"]], on="timestamp")
        # Power consumption
        tegra_total_power = df[(df["name"] == "tegra_wattage_current_milliwatts") & (df["meta.sensor"] == "vdd_in")]
        print(f'{len(tegra_total_power)} power metric records found')
        # We filter out all measurements
        tegra_total_power = tegra_total_power.rename({"value": "system_power"}, axis="columns")
        merged_instance = pd.merge_asof(merged_instance, tegra_total_power[["timestamp", "system_power"]], on="timestamp")

        tegra_cpugpu_power = df[(df["name"] == "tegra_wattage_current_milliwatts") & (df["meta.sensor"] == "vdd_cpu_gpu_cv")]
        tegra_cpugpu_power = tegra_cpugpu_power.rename({"value": "cpugpu_power"}, axis="columns")
        merged_instance = pd.merge_asof(merged_instance, tegra_cpugpu_power[["timestamp", "cpugpu_power"]], on="timestamp")

        merged_instance["plugin_instance"] = instance
        merged_instance['timestamp'] = merged_instance['timestamp'].map(lambda x: x.isoformat())
        merged = pd.concat([merged, merged_instance])
    merged.to_csv(f'{plugin_name}_{vsn}.csv', index=False)
````
